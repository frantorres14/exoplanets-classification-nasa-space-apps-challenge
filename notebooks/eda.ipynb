### K2
default_flag: Indica si este es el registro principal del planeta.
sy_snum: Número de estrellas en el sistema.
sy_pnum: Número de planetas en el sistema.
sy_mnum: Número de lunas detectadas (si se conocen).
cb_flag: El planeta orbita un sistema binario.
disc_year: Año del descubrimiento del planeta.
rv_flag: Detectado por velocidad radial (el planeta tira de la estrella).
pul_flag: Detectado por variaciones en un púlsar (estrella muerta que pulsa).
ptv_flag: Detectado por cambios en el brillo de la estrella.
tran_flag: Detectado porque el planeta pasó frente a su estrella (tránsito).
ast_flag: Detectado por astrometría (movimiento de la estrella en el cielo).
obm_flag: Detectado con imágenes directas.
micro_flag: Detectado por microlente gravitacional (efecto de aumento de luz).
etv_flag: Detectado por cambios en el tiempo de eclipses.
ima_flag: Confirmado con imágenes de alta resolución.
dkin_flag: Confirmado por análisis dinámico del sistema.
pl_controv_flag: Hay dudas sobre si realmente es un planeta.
pl_orbper: Tiempo que tarda en dar una vuelta completa a su estrella (días).
pl_tranmid: Momento central del último tránsito observado.
ttv_flag: Presenta cambios en el tiempo de sus tránsitos (posible interacción con otro planeta).
st_rad: Tamaño de la estrella (en radios solares).
ra: Posición de la estrella en el cielo (ascensión recta).
dec: Posición de la estrella en el cielo (declinación).
glat: Latitud galáctica (posición en la Vía Láctea).
glon: Longitud galáctica (posición en la Vía Láctea).
elat: Latitud en el plano del sistema solar.
elon: Longitud en el plano del sistema solar.
sy_pm: Velocidad con la que se mueve la estrella en el cielo.
sy_pmra: Movimiento de la estrella en ascensión recta.
sy_pmdec: Movimiento de la estrella en declinación.
sy_dist: Distancia al sistema (en parsecs).
sy_bmag: Brillo de la estrella en luz azul.
sy_vmag: Brillo en luz visible.
sy_jmag: Brillo en infrarrojo cercano (banda J).
sy_hmag: Brillo en infrarrojo cercano (banda H).
sy_kmag: Brillo en infrarrojo cercano (banda K).
sy_w1mag: Brillo en infrarrojo medio (banda W1).
sy_w2mag: Brillo en infrarrojo medio (banda W2).
sy_w3mag: Brillo en infrarrojo medio (banda W3).
sy_w4mag: Brillo en infrarrojo medio (banda W4).
sy_gaiamag: Brillo medido por la misión Gaia.
sy_tmag: Brillo en banda TESS.
sy_kepmag: Brillo en banda Kepler.
pl_nnotes: Número de notas o comentarios asociados al planeta.
k2_campaigns_num: Número de campañas de observación de K2 en las que aparece.
st_nphot: Número de observaciones de brillo de la estrella.
st_nrvc: Número de mediciones de velocidad radial de la estrella.
st_nspec: Número total de espectros tomados de la estrella.
pl_nespec: Número de espectros donde aparece el planeta.
pl_ntranspec: Número de espectros tomados durante un tránsito.
pl_ndispec: Número de espectros tomados fuera del tránsito.

### KOI
kepid: Identificador único de la estrella según el catálogo Kepler.
koi_fpflag_nt: Marcado como falso positivo por no mostrar un tránsito real.
koi_fpflag_ss: Marcado como falso positivo por ser probablemente una estrella binaria o sistema estelar.
koi_fpflag_co: Marcado como falso positivo por contaminación de otra fuente de luz cercana.
koi_fpflag_ec: Marcado como falso positivo por tener señales parecidas a un eclipse estelar.
koi_period: Tiempo que tarda el objeto en orbitar su estrella (días).
koi_time0bk: Momento del tránsito en tiempo juliano corregido para la órbita de Kepler.
koi_time0: Momento del tránsito en tiempo juliano estándar.
koi_eccen: Excentricidad de la órbita (qué tan circular o elíptica es).
koi_impact: Qué tan centrado pasa el planeta frente a la estrella.
koi_duration: Duración del tránsito (horas).
koi_depth: Profundidad del tránsito (cuánto baja el brillo de la estrella).
koi_ror: Tamaño del planeta comparado con el tamaño de la estrella.
koi_srho: Densidad estimada de la estrella.
koi_prad: Tamaño estimado del planeta (en radios terrestres).
koi_sma: Distancia media entre el planeta y la estrella.
koi_incl: Inclinación de la órbita con respecto a la vista desde la Tierra.
koi_teq: Temperatura estimada del planeta.
koi_insol: Cantidad de luz que recibe el planeta comparado con la Tierra.
koi_dor: Distancia entre el planeta y la estrella dividida por el radio de la estrella.
koi_ldm_coeff4: Coeficiente 4 del modelo de oscurecimiento del borde estelar.
koi_ldm_coeff3: Coeficiente 3 del modelo de oscurecimiento del borde estelar.
koi_ldm_coeff2: Coeficiente 2 del modelo de oscurecimiento del borde estelar.
koi_ldm_coeff1: Coeficiente 1 del modelo de oscurecimiento del borde estelar.
koi_model_snr: Calidad de la señal del tránsito (señal/ruido).
koi_count: Número de tránsitos observados.
koi_tce_plnt_num: Número de planeta candidato dentro del sistema.
koi_steff: Temperatura de la estrella.
koi_slogg: Gravedad superficial de la estrella.
koi_smet: Metalicidad de la estrella (qué tantos elementos pesados tiene).
koi_srad: Radio de la estrella.
koi_smass: Masa de la estrella.
ra: Posición de la estrella en el cielo (ascensión recta).
dec: Posición de la estrella en el cielo (declinación).
koi_kepmag: Brillo de la estrella en banda Kepler.
koi_gmag: Brillo de la estrella en banda Gaia G.
koi_rmag: Brillo de la estrella en banda roja.
koi_imag: Brillo de la estrella en banda infrarroja I.
koi_zmag: Brillo de la estrella en banda Z.
koi_jmag: Brillo en infrarrojo cercano (banda J).
koi_hmag: Brillo en infrarrojo cercano (banda H).
koi_kmag: Brillo en infrarrojo cercano (banda K).
koi_fwm_stat_sig: Significancia estadística del análisis de imagen.
koi_fwm_sra: Ajuste en ascensión recta de la fuente principal.
koi_fwm_sdec: Ajuste en declinación de la fuente principal.
koi_fwm_srao: Posición original en ascensión recta.
koi_fwm_sdeco: Posición original en declinación.
koi_fwm_prao: Posición del posible planeta en ascensión recta.
koi_fwm_pdeco: Posición del posible planeta en declinación.
koi_dicco_mra: Diferencia entre métodos en ascensión recta.
koi_dicco_mdec: Diferencia entre métodos en declinación.
koi_dicco_msky: Diferencia total en posición en el cielo.
koi_dikco_mra: Diferencia en ascensión recta por contaminación.
koi_dikco_mdec: Diferencia en declinación por contaminación.
koi_dikco_msky: Diferencia total en posición por contaminación.


### TOI
toi: Identificador único del objeto de interés detectado por TESS.
toipfx: Versión o prefijo del identificador TOI (variante del catálogo).
tid: Identificador de la estrella en el catálogo TIC (TESS Input Catalog).
ctoi_alias: Alias alternativo del TOI (otro nombre con el que puede aparecer).
pl_pnum: Número del planeta dentro del sistema (ej. 1 si es el primero, 2 si hay más).
ra: Posición de la estrella en el cielo (ascensión recta).
dec: Posición de la estrella en el cielo (declinación).
st_pmra: Movimiento de la estrella en ascensión recta (cómo se mueve en el cielo).
st_pmdec: Movimiento de la estrella en declinación.
pl_tranmid: Momento central del tránsito más reciente.
pl_orbper: Tiempo que tarda el planeta en completar una órbita (días).
pl_trandurh: Duración del tránsito en horas (cuánto tiempo tapa la estrella).
pl_trandep: Profundidad del tránsito (cuánto baja el brillo de la estrella).
pl_rade: Tamaño estimado del planeta en radios terrestres.
pl_insol: Cantidad de luz que recibe el planeta comparado con la Tierra.
pl_eqt: Temperatura de equilibrio del planeta (su temperatura estimada).
st_tmag: Brillo de la estrella en la banda TESS.
st_dist: Distancia de la estrella a la Tierra (en parsecs).
st_teff: Temperatura de la estrella.
st_rad: Radio de la estrella (en radios solares).
